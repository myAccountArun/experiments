Time series with quadratic function

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# create a quadratic function for variable x
def f(x):
    #get white noise for x
    mean = 2
    std = 25
    noise = np.random.normal(mean, std, len(x))
    # create a quadratic function with noise
    arrayX = 0.2*x**2 + 0.1*x + 2 + noise
    return arrayX
# create a range of x values
x = np.linspace(0, 100, 100).astype(int)
print(x)
# create a DataFrame with x values and their corresponding f(x) values
y = f(x)
# plot the function
df = pd.DataFrame({'x': x, 'y': y})
# save the DataFrame to a CSV file
df.to_csv('quadratic_function.csv', index=False)
```

# Import dataset for Analysis

```python
# Load quadratic function data from CSV
df = pd.read_csv('quadratic_function.csv')

# Sort df on x values
df = df.sort_values(by='x')
```

```python
# Plot the quadratic function
plt.figure(figsize=(10, 6))
plt.plot(df['x'], df['y'], label='y = 0.2*x**2 + 0.1*x + 2 + noise', color='blue')
plt.title('Quadratic Function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.legend()
# plt.savefig('quadratic_function_plot.png')
plt.show()
```

First Difference

```python
# Create a function to calcualte first difference
def first_difference(df):
    # Calculate the first difference
    df['first_difference'] = df['y'].diff()[1:]
    # Calculate the second difference
    df['second_difference'] = df['first_difference'].diff()[2:]
    return df
```

```python
df_diff = first_difference(df)
# Plot the first difference
plt.figure(figsize=(10, 6))
plt.plot(df_diff['x'], df_diff['first_difference'], label='First Difference', color='red')
plt.plot(df_diff['x'], df_diff['second_difference'], label='Second Difference', color='green')
# plt.fill_between(df_diff['x'], df_diff['first_difference'], color='red', alpha=0.2)
# plt.fill_between(df_diff['x'], df_diff['second_difference'], color='green', alpha=0.2)
plt.title('First and Second Differences of Quadratic Function')
plt.xlabel('x')
plt.ylabel('Difference')
plt.grid()
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.legend()
plt.show()
```
